Data Overview:

My project consists of 2 data sets derived from Kaggle. The first is contains about 175,000 songs released between 1921 and 2020. This first data set was used to find the trends amongst features of music over time. The second data set was used to look at the genres that coincide with the muscial features observed.

Hypothesis: Music is made up of a variety of elements, a popular genre of music will rank high amongst multiple musical features. High trending musical features will also be related. 

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data 2.csv')
df
#We read the csv file into a pandas dataframe.
```

```python
df.info()
#So we can see the data set contains 174,389 entries, 19 columns, andthe data types of each entry.
```

```python
df.isnull().sum()
#The data set contains no null values.
```

```python
df.groupby(['year']).count()
#To see how many entries wehave We see that there are duplicates. 
```

```python
music_trends_by_year = df.groupby(['year'],as_index = False).mean()
music_trends_by_year
#I decided to group by the year released, then take the mean which returned the mean of each, also make not that 
#Acousticness, Danceability, Energy, Explicit, Liveness, Popularity, Valence share the same scale 
```

```python
#Here is where I wanted to look at some of the features in individual graphs. 
```

```python
plt.plot('year', 'danceability',data = music_trends_by_year)
plt.title('Music Trends')
plt.xlabel('year')
plt.ylabel('danceability')
plt.show()
```

```python
plt.plot('year', 'acousticness',data = music_trends_by_year)
plt.title('Music Trends')
plt.xlabel('year')
plt.ylabel('acousticness')
plt.show()
```

```python
plt.plot('year', 'energy',data = music_trends_by_year)
plt.title('Music Trends')
plt.xlabel('year')
plt.ylabel('energy')
plt.show()
```

```python
# I decided to plot popularity as a bar graph because a line graph would be less accurate, bc its actually static vs over time.
```

```python
music_trends_by_year.plot.bar('year','popularity')
plt.rcParams["figure.figsize"] = (20,15)
plt.xlabel('Year')
plt.ylabel('Values')
plt.title('Popularity Over The Years')
```

```python
music_trends_by_year.sort_values(by='popularity', ascending=False)[['year','popularity']][:10]
```

```python
#Then I wanted to see the line graphs aggregated to see any interesting trends. 
```

```python
fig,ax = plt.subplots(figsize=(20,8))
ax.set_title('Music Trends over Time')
ax.plot(music_trends_by_year.year, music_trends_by_year.danceability, label = 'Danceability')
ax.plot(music_trends_by_year.year, music_trends_by_year.acousticness, label = 'Acousticness')
ax.plot(music_trends_by_year.year, music_trends_by_year.energy, label = 'Energy')
ax.plot(music_trends_by_year.year, music_trends_by_year.valence, label = 'Valence')
ax.plot(music_trends_by_year.year, music_trends_by_year.liveness, label = 'Liveness')
ax.plot(music_trends_by_year.year, music_trends_by_year.explicit, label = 'Explicit')
ax.legend()
plt.xlabel('Year')
plt.ylabel('Values')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df_top = pd.read_csv('top10s.csv', encoding='mac_roman')
df_top
```

```python
df_top.info()
```

```python
df_top.isnull().sum()
```

```python
df_top.groupby(['top genre']).mean()
```

```python
reindexed_df_top = df_top.groupby(['top genre'],as_index = False).mean()
reindexed_df_top

#We need to reindex 
```

```python
danceability = reindexed_df_top.sort_values(by='dnce', ascending=False)[['top genre','dnce']][:10]
danceability
#Taking a look at avg danceability by genre from max to the lowest value. 
```

```python
energy = reindexed_df_top.sort_values(by='nrgy', ascending=False)[['top genre','nrgy']][:10]
energy
#Taking a look at avg energy by genre from max to the lowest value.
```

```python
valence = reindexed_df_top.sort_values(by='val', ascending=False)[['top genre','val']][:10]
valence
#Taking a look at avg valence by genre from max to the lowest value.
```

```python
popularity = reindexed_df_top.sort_values(by='pop', ascending=False)[['top genre','pop']][:10]
popularity
```

```python
df_top.groupby(['top genre']).count().sort_values(by='pop', ascending=False)
#We see that there are discrepancies based on how many entries there are per genre. The data is severely skewed.
```

```python
corr = df_top[['acous','dnce','nrgy',
'live','val']].corr()
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
%matplotlib inline
plt.figure(figsize=(12,8))
plt.title('Correlation Between Musical Features')
sns.heatmap(corr, annot=True)
```

We see that danceability and acousticness are not closely related, but it has some relation to energy and valence,which supports the hypothesis of trending musical elements being closely related. However, there aren't enough entries to take a look at the genres in depth.
